```python
import requests
import pandas as pd #bibléotheque qui fait du trautelent de données
from datetime import datetime
#Extraction des données via API
url = "https://gbfs.api.ridedott.com/public/v2/lyon/free_bike_status.json"

response = requests.get(url)
data = response.json()

bikes = data["data"]["bikes"]
df = pd.DataFrame(bikes)

# Ajouter timestamp d’extraction
df["timestamp"] = datetime.now().isoformat()

df.to_csv("dott_bikesh.csv", index=False)
```

```python
import folium
from folium.plugins import HeatMap

map_center = [df['lat'].mean(), df['lon'].mean()]
m = folium.Map(location=map_center, zoom_start=100)

HeatMap(data=df[['lat', 'lon']].dropna().values.tolist()).add_to(m)
m
```

```python
import requests
import pandas as pd
from datetime import datetime

# New URL for station information
url = "https://gbfs.api.ridedott.com/public/v2/lyon/station_information.json"

# Fetch the data from the API
try:
    response = requests.get(url)
    response.raise_for_status() # Raise an exception for bad status codes (4xx or 5xx)
    data = response.json()
except requests.exceptions.RequestException as e:
    print(f"Error fetching data from API: {e}")
    data = None # Set data to None if there's an error

if data and "data" in data and "stations" in data["data"]:
    # The data for station information is usually under a key like "stations"
    stations = data["data"]["stations"]
    df_stations = pd.DataFrame(stations)

    # Ajouter timestamp d’extraction
    df_stations["timestamp"] = datetime.now().isoformat()

    # Sauvegarder dans un fichier CSV
    # Using a different filename to distinguish from free bike status
    df_stations.to_csv("dott_station_information.csv", index=False)

    print("Station information successfully fetched and saved to dott_station_information.csv")
    # Display the first few rows and data types for verification
    display(df_stations.head().T)
    print("\nDataFrame Info:")
    df_stations.info()

else:
    print("Could not retrieve station information data from the API response.")
```

```python
map_center = [df_stations['lat'].mean(), df_stations['lon'].mean()]
m = folium.Map(location=map_center, zoom_start=100)

HeatMap(data=df_stations[['lat', 'lon']].dropna().values.tolist()).add_to(m)
m
```
